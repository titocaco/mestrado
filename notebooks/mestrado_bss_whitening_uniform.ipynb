```python
import scipy.stats as ss
import matplotlib.pyplot as plt
import numpy as np
```

```python
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [5, 5],
'axes.labelsize': 22,
'axes.titlesize':26,
'font.size': 16,
'legend.fontsize': 12,
'xtick.labelsize': 18,
'ytick.labelsize': 18
}
plt.rcParams.update(params)
# plt.style.use('seaborn-notebook')
```

```python
data1 = ss.uniform.rvs(loc=-1.75, scale=3.50, size = 2000, random_state=None)
data2 = ss.uniform.rvs(loc=-1.75, scale=3.50, size = 2000, random_state=None)

data = np.array([[data1[i], data2[i]] for i in range(data1.shape[0])])
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data[:, 0], data[:, 1], edgecolors='w', alpha=0.8)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-4, -2, 0, 2, 4])
plt.grid(True)
plt.xlabel('x1')
plt.ylabel('x2')
plt.savefig('bss_whitening_uniform.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
a1 = 0.3
a2 = 0.9

obs_data = np.array([[a1 * data1[i] + a2 * data2[i], 
                      a2 * data1[i] + a1 * data2[i]] for i in range(data1.shape[0])])
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(obs_data[:, 0], obs_data[:, 1], edgecolors='w', alpha=0.8)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-4, -2, 0, 2, 4])
plt.grid(True)
plt.xlabel('x1')
plt.ylabel('x2')
plt.savefig('bss_whitening_uniform_mixture.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(whiten=True)

whitened_obs_data = pca.fit_transform(obs_data)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(whitened_obs_data[:, 0], whitened_obs_data[:, 1], edgecolors='w', alpha=0.8)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-4, -2, 0, 2, 4])
plt.grid(True)
plt.xlabel('x1')
plt.ylabel('x2')
plt.savefig('bss_whitening_uniform_mixture_whitened.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
